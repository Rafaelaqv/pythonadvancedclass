```python
##Self-test 12-11


# example: (time=4, elevation=2, latitude=3, longitude=2)
humidity = np.random.rand(4, 2, 3, 2)

# difference between consecutive time steps
diff = humidity[1:,:,:,:] - humidity[0:-1,:,:,:]

print("Original shape:", humidity.shape)
print("Diff shape:", diff.shape)

```

```python
##Self-test 12-12

# Example array: (time=4, elevation=2, latitude=3, longitude=2)
humidity = np.random.rand(4, 2, 3, 2)

# Define the shape of the diff array (time-1, elevation, latitude, longitude)
diff_shape = (humidity.shape[0] - 1, humidity.shape[1], humidity.shape[2], humidity.shape[3])
diff = np.zeros(diff_shape)

# Loop through each dimension
for i in range(diff_shape[0]):        # time (except the last one)
    for j in range(diff_shape[1]):    # elevation
        for k in range(diff_shape[2]):# latitude
            for l in range(diff_shape[3]): # longitude
                diff[i,j,k,l] = humidity[i+1,j,k,l] - humidity[i,j,k,l]

print("Original shape:", humidity.shape)
print("Diff shape:", diff.shape)
```

```python
##Self-test 12-13

# Example humidity array: (time=4, elevation=2, latitude=3, longitude=2)
humidity = np.random.rand(4, 2, 3, 2)

# Example process_humidity function: here it just returns the mean value
def process_humidity(subarray):
    return np.mean(subarray)

# Prepare array to store results
processed = np.zeros(humidity.shape[0])

# Loop through each time slice
for t in range(humidity.shape[0]):
    subarray = humidity[t, :, :, :]       # select elevation-latitude-longitude subarray
    processed[t] = process_humidity(subarray)

print("Processed results:", processed)
print("Length of processed:", len(processed))
```

**14-1. Describe the differences between a Figure, Axes, and Line2D object.**
A Figure is the overall container that holds everything in a visualization, representing the page or window on which one or more graphs appear. An Axes object is a specific region inside the Figure where the actual data is drawn, including the x- and y-axes, and it represents a single graph or plot. A Line2D object represents the individual line or curve that is drawn within an Axes, based on the data values provided.


**14-2. What do pyplot functions subplot and subplots do and how do they differ from each other?**
subplot and subplots create Axes object(s) within the context of a single or multipanel Figure. The subplot function creates one Axes object inside a Figure, usually specifying its position in a grid layout. In contrast, the subplots function can generate multiple Axes objects at once, arranged in rows and columns, and it returns both the Figure and the collection of Axes. In other words, subplot is used for creating one panel at a time, whereas subplots is more convenient for generating a grid of panels in a single call.
14-3. What is the difference between normalized and data coordinates?
Both coordinate systems are ways of describing a location in the plot frame. Normalized coordinates describe a location using a scale from 0 to 1, where 0 represents the start of an axis and 1 represents the end. They are relative to the size of the plot frame. Data coordinates, on the other hand, are tied to the actual values being plotted, ranging from the minimum to the maximum data values on the axes. For example, a value at 0.5 in normalized coordinates would mean the midpoint of the axis, but in data coordinates that same position could correspond to something like year 2010 on an axis spanning 2000–2020.

**14-4. List and describe some Axes object methods:**
Some useful methods of an Axes object include:
*	plot() – draws line plots of x and y data values. 
*	contour() – creates contour line plots that connect points with equal values. 
*	contour() – similar to contour, but fills the spaces between contour lines with colors. 
*	imshow() – displays an image array on the axes. 
*	set_xlabel() / set_ylabel() – add labels to the x- and y-axes. 
*	set_title() – gives the plot a title.


**14-5. What are some circumstances to use the Matplotlib object API instead of the pyplot interface?**
The Matplotlib object-oriented API is useful when we need more precise control over a figure. For example, if we want to place multiple plots in custom position, combine different types of plots on the same axes, or adjust elements like labels, ticks, or colorbars in a detailed way. In contrast, the pyplot interface is quicker for simple plots, but the object API is better when building complext or highly customized visualizations. 

**14-6. Define a contour line plot and a filled contour plot. What are the Matplotlib commands or methods to create these kinds of plots?**
A countor line plot shows only the contour lines that connect points of equal value, similar to elevation lines on a map. A filled contour plot, on the other hand, shades the spces between contour lines to highlight the ranges of values. In Marplotlib, the contour method is used to create contour line plots, while the contourf is used to create filled contour plots.  


**14-7. What are some ways of setting the color map for a filled contour plot?**
All of the methods work by setting the cmap keyword argument in the call to contourf. There are several ways to do this:
*	By giving the name of a predefined colormap as a string (e.g., “viridis”, “plasma”). 
*	By passing a colormap object from the matplotlib.cm module. 
*	By using the return value of plt.get.cmap() with the chosen colormap name. 
*	By creating a custom colormap object (for example, with LinearSegnmentedColormap).

**14-8. Summarize the steps to creating a contour line plot or filled contour plot overlain on a map.**
**1.**	Create the Axes with a map projection – define an Axes object and set the projection keyword to choose the map projection (e.g., Lambert, PlateCarree). 
**2.**	Customize the map background – add features such as coastlines, borders, or gridlines to give geographic context. 
**3.**	Plot the data with contour/contour – call contour or contourf to create the contour lines or filled contours from the dataset. 
**4.**	Specify the transform – use the transform parameter to tell Matplotlib/Cartopy the coordinate system of your data, so it can be correctly placed on the chosen map projection. 

**14-9. What are the steps to creating a basic animation of images? Of a line plot?**
To create a basic animation, we generally set up a loop that updates the plot at each step.
* For images: we begin by displaying the first frame using the imshow method of an Axes object. As the loop runs, the image is updated with the set_data method applied to the AxesImage object.
* For line plots: we start by creating the initial curve with the plot method of an Axes object. In later iterations, the line is updated using the set_xdata and set_ydata methods of the Line2D object returned by the initial plot call.
* In both cases: after the content is updated, we call draw to render the new frame and pause to briefly stop the loop so that the animation progresses smoothly.


```python
#Self-Question 14-10
import numpy as np 
import matplotlib.pyplot as plt

# Dictionary containing data and metadata
data = {
    'hour_in_day': np.array([9, 11, 13, 15, 17]),   # Hours of the day (9 = 9:00 a.m., 13 = 1:00 p.m.)
    'bird_count': np.array([3, 5, 2, 2, 4]),       # Bird counts at each hour
    'rabbit_count': np.array([2, 1, 0, 0, 3]),     # Rabbit counts at each hour
    'location': '123 Main St.',                     # Metadata: location of observation
    'date': 'May 1, 2020'                           # Metadata: date of observation
}

plt.figure()  # Create a new figure window

# Plot bird counts with red circles and solid line
plt.plot(data['hour_in_day'], data['bird_count'], 'ro-', label="Birds")

# Plot rabbit counts with blue squares and dashed line
plt.plot(data['hour_in_day'], data['rabbit_count'], 'bs--', label="Rabbits")

# Add a title using metadata (location and date)
plt.title('Animal Counts at ' + data['location'] + ' on ' + data['date'])

# Label the x-axis (hour of the day)
plt.xlabel('Hour in Day')

# Label the y-axis (counts of animals)
plt.ylabel('Counts')

# Add a legend to distinguish birds from rabbits
plt.legend()

plt.show()  # Display the plot
```
