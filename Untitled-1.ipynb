```python
# Wind speed measured at different hours (m/s)
velocity = np.array([3.5, 5.2, 7.8, 2.1, 9.3, 6.5, 4.0, 10.2])

# Define "strong gust" as any wind speed above 6 m/s
threshold = 6.0

sum_over_threshold = 0.0
count = 0
for ivalue in velocity:
    if ivalue > threshold:
        sum_over_threshold += ivalue
        count += 1
over_threshold_avg = sum_over_threshold / float(count)

print("Values above threshold:", velocity[velocity > threshold])
print("Average of strong gusts:", over_threshold_avg)
```
