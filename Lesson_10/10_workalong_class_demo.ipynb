```python
""" Define a class called Animal. """

class Animal:

    def __init__(self, limbs = 4, eyes = 2, tails = 1):
        """ This is the constructor: it is called when a new object of type 'Animal' is created. """
        self.limbs = limbs
        self.eyes = eyes
        self.tails = tails
        print("Init has been called!")

    # define a method called "speak"
    def speak(self):
        """This is a method: it is a function that is associated with an object."""
        print("Hello, I am an animal!")
        return self # enable method chaining
    
    def describe(self):
        """ This is another method that describes the animal"""
        print(f"I have {self.limbs} limbs, {self.eyes}, eyes, and {self.tails} tail(s).")
        return self # enable method chaining
```

```python
""" Create two animal objects."""
animal1 = Animal() # instantiate an object from Animal

animal2 = Animal(limbs = 2, eyes = 2, tails = 0) # instantiate another animal with different attribute
```

```python
""" Call the methods of the two animal objects. """

print("Animal 1:")
animal1.speak()
animal1.describe()
print()

print("Animal 2:")
animal2.speak()
animal2.describe()
print()
```

```python
""" Demonstrate method chaining """
animal1.speak().describe();
```

# Create a class from another class

```python
class Dog(Animal):
    def __init__(self, name, breed):
        # initialize the parent class, but force the number of limbs and tails
        super().__init__(limbs=4, eyes = 2, tails=1)

        # store the attributes
        self.name = name
        self.breed = breed
    
    def speak(self):
        """ This method overloads the parent's speak() method."""
        print(f"{self.name}: Woof!")
        return self
    
    def __repr__(self):
        """ This is a special metod that is called when the object is printed. """
        return f"Dog(name= '{self.name}, breed = '{self.breed})"


dog1 = Dog(name = "Gabby", breed = "corgi")
dog1.speak().describe(); 
```

# Create a Quadratic class
 This class will represent a quadratic polynomial of the form axÂ² + bx +c
 
 The quadratic has the derivate 

```python
class Quadratic:
    
    def __init__(self, a, b, c):
        """ This is a constructor; it is called when a Quadratic is created. """
        self.a = a
        self.b = b
        self.c = c
    
    def evaluate(self, x):
        
        return self.a * x**2 + self.b * x + self.c
    
    def __repr__(self):
        """ Print the quadratic. """
        return f"{self.a} * x **2 + {self.b} * x + {self.c}"
    
    def __call__(self, x):
        """ Evaluate the quadratic, treating the object like a function. """
        return self.evaluate(x)   

    def gradient(self, x):
        """ Calculate the gradient of the quadratic at a given point. """ 
        return 2*self.a*x + self.b
```

```python
quad1 = Quadratic(0, 2, 3)

quad1
```
