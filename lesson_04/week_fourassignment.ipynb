```python
##first way##

data3 = np.array([[3, 8, 2],
                  [10, 5, 7],
                  [6, 4, 9]])

max_data3 = np.max(data3)
print("Max. with NumPy:", max_data3)

##second way#3

max_data3=data3[0,0]
for i in range(data3.shape[0]):        # loop through the rows
    for j in range(data3.shape[1]):    # loop through the columns
        if data3[i, j] > max_data3:    # compare current element
            max_data3 = data3[i, j]    # update the maximum

print("Max. with loops:", max_data3)
```
